```python
import os

from core.tools import read_session_note, save_session_note

LOCAL_STORE_URL = os.environ.get(
    "MDCROW_STORE_DATABASE_URL",
    "postgresql://postgres:postgres@localhost:5432/postgres?sslmode=disable",
)

ltm = MDCrow(
    model=llm_var,
    store_conn_string=LOCAL_STORE_URL,
    extra_tools=(save_session_note, read_session_note),
)
```

```python
# Ask the model to write into long-term memory (store.put via save_session_note)
ltm.run("Please remember: my favorite PDB id for testing is 1LYZ. Call save_session_note with a short line stating that fact.")
```

```python
# New user message only - short-term may not include the last turn if you rely only on the graph;
# read_session_note loads from the store, so the fact should still appear.
ltm.run("I forgot what I told you. What is my favorite PDB id for testing? Use read_session_note.")
```
